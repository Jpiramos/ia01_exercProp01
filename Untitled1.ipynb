```python
import torch
from torch.utils.data import Dataset, DataLoader
import numpy as np
import matplotlib.pyplot as plt

# 1. Criando o Dataset
class CustomDataset(Dataset):
    def __init__(self, n_samples=100):
        self.X = np.random.rand(n_samples, 1) * 10  # Inputs aleatórios
        self.y = 2 * self.X + 3  # Relação linear simples (y = 2X + 3)

        # Convertendo para tensores do PyTorch
        self.X = torch.tensor(self.X, dtype=torch.float32)
        self.y = torch.tensor(self.y, dtype=torch.float32)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, index):
        return self.X[index], self.y[index]

# Função para plotar o dataset
def plot_dataset(dataset):
    X, y = dataset.X.numpy(), dataset.y.numpy()
    plt.scatter(X, y, label='Dataset Original')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.title('Relação Linear entre X e y')
    plt.legend()
    plt.show()

# Criando o dataset e o DataLoader
dataset = CustomDataset(n_samples=100)
dataloader = DataLoader(dataset, batch_size=10, shuffle=True)

# Plotando o dataset gerado
plot_dataset(dataset)

```

```python
import torch.nn as nn

# 2. Definindo o Modelo da Rede Neural
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(1, 64)  # Camada de entrada com 64 neurônios
        self.fc2 = nn.Linear(64, 32)  # Primeira camada escondida
        self.fc3 = nn.Linear(32, 1)   # Camada de saída

    def forward(self, x):
        x = torch.relu(self.fc1(x))  # ReLU na primeira camada
        x = torch.relu(self.fc2(x))  # ReLU na segunda camada
        x = self.fc3(x)              # Saída linear
        return x

# Instanciando o modelo
model = SimpleNN()

```

```python
# 3. Definindo a Função de Perda e o Otimizador
criterion = nn.MSELoss()  # Mean Squared Error Loss
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)  # Adam optimizer

# Lista para salvar os valores de perda ao longo do treinamento
losses = []

```

```python
# 4. Treinamento do Modelo
n_epochs = 100  # Número de épocas
for epoch in range(n_epochs):
    epoch_loss = 0.0
    for inputs, targets in dataloader:
        # Forward pass
        outputs = model(inputs)
        loss = criterion(outputs, targets)

        # Backward pass e otimização
        optimizer.zero_grad()  # Zerar gradientes
        loss.backward()  # Retropropagação
        optimizer.step()  # Atualizar os pesos

        epoch_loss += loss.item()

    # Salvando a perda média por época
    losses.append(epoch_loss / len(dataloader))

    # Exibir a perda a cada 10 épocas
    if (epoch+1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{n_epochs}], Loss: {epoch_loss:.4f}')
```

```python
# Plotando a curva de perda
plt.plot(range(1, n_epochs+1), losses, label='Perda')
plt.xlabel('Épocas')
plt.ylabel('Perda')
plt.title('Curva de Perda Durante o Treinamento')
plt.legend()
plt.show()

```

```python
# 5. Usando o modelo treinado para previsões
with torch.no_grad():
    new_input = torch.tensor([[4.5]])  # Exemplo de entrada nova
    prediction = model(new_input)
    print(f'Input: 4.5, Predicted Output: {prediction.item():.4f}')
```
